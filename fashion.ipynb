```python
import tensorflow as tf
import numpy as np
from tensorflow import keras

print(tf.__version__)
```

```python
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer='sgd', loss='mean_squared_error')
xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
model.fit(xs, ys, epochs=50)
```

```python
print(model.predict([10.0]))
```

```python
fashiondata=tf.keras.datasets.fashion_mnist
(t_img,t_leb) ,(test_img,test_leb)=fashiondata.load_data()
```

```python
import numpy as np
import matplotlib.pyplot as plt
np.set_printoptions(linewidth=320)
print(f"LABEL:{t_img[90]}")
print(f"\n Images Pixel Array\n")
plt.imshow(t_img[90])
```

```python
t_img=t_img/255.0
test_img=test_img/225.0
```

```python
model = tf.keras.models.Sequential([tf.keras.layers.Flatten(), 
                                    tf.keras.layers.Dense(256, activation=tf.nn.relu), 
                                    tf.keras.layers.Dense(10, activation=tf.nn.softmax)])
```

```python
model.compile(optimizer = tf.optimizers.Adam(),
              loss = 'sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(t_img, t_leb, epochs=100)
```

```python
model.evaluate(test_img,test_leb)
```
